# Boosting

```python
train3 = train
train3 = train3.drop(['Name','Survived'], axis =1)
test3 = test
test3 = test3.drop(['Name'], axis =1)
train3.head()
```

```python
train3['Sex'] = train3['Sex'].replace({'male':0, 'female':1})
train3['Ticket'] = train3['Ticket'].replace("LINE", 0)
train3['Ticket'] = train3['Ticket'].str.split(' ').str[-1].astype('float')
train3['Embarked'] = train3['Embarked'].replace({'S':0, 'C':1, 'Q':2})
train3.head()
```

```python
test3['Sex'] = test3['Sex'].replace({'male':0, 'female':1})
test3['Ticket'] = test3['Ticket'].str.split(' ').str[-1].astype('int')
test3['Embarked'] = test3['Embarked'].replace({'S':0, 'C':1, 'Q':2})
test3.head()
```

```python
train3['Cabin'].unique()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cabin_encoding = pd.Series([i for i in train3['Cabin'] if type(i) == str])
le.fit(cabin_encoding)
train3['Cabing_encoding'] =cabin_encoding.apply(lambda x : le.transform([x])[0] if type(x) == str else x)
train3.head()
```

```python
test3['Cabin'].unique()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cabin_encoding = pd.Series([i for i in test3['Cabin'] if type(i) == str])
le.fit(cabin_encoding)
test3['Cabing_encoding'] =cabin_encoding.apply(lambda x : le.transform([x])[0] if type(x) == str else x)
test3.head()
```

```python
train3 = train3.drop('Cabin',axis =1 )
test3 = test3.drop('Cabin', axis =1)
```
